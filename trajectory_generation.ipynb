```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/home/danielriege/Desktop/01.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
height, width, _ = image.shape

print(width)

plt.figure(figsize=(8, 6), dpi=300)
plt.imshow(image)
plt.show()
```

```python
xs = []
ys = []
for x in range(width):
    for y in range(height):
        if image[y,x,2] > 150:
            xs.append(x)
            ys.append(y)

z = np.polyfit(xs,ys,3)
print(z)
p = np.poly1d(z)

xp = np.linspace(0, 208, 208)
plt.plot(xs,ys,'.',xp,p(xp), '-')
plt.ylim(0,height)
```
